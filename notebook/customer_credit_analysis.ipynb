```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_clean = pd.read_csv("../data/credit_clients_clean.csv")
data_clean.head()
```

```python
print("__________________________________________________________________________________________________________________________")
print("ANALYSIS")

#correlacón entre las variables creadas y el limite crediticio
corr = data_clean[
    ["LIMIT_BAL","TOTAL_BILL","TOTAL_PAY","DEBT_RATIO"]
].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Between Credit Variables")

plt.show()
```

```python

#proporcion de clientes que hacen default, 0=si pago deuda  1=no pago deuda

print("Distribution of Credit Card Default")
data_clean["default payment next month"].value_counts(normalize=True)

sns.countplot(x="default payment next month", data=data_clean)
plt.title("Distribution of Credit Card Default")
plt.xlabel("Default Next Month")
plt.ylabel("Number of Clients")
plt.xticks([0,1],["Payd","Defaulted"])

plt.show()
```

```python
print("Default Rate by Credit Limit Segment")

#vamos a crear limites para armar los grupos
bins = [0,50000,150000,300000,1000000]

labels = ["Very Low Limit",
          "Low Limit",
          "Medium Limit",
          "High Limit"]

data_clean["limit_segment"] = pd.cut(
    data_clean["LIMIT_BAL"],
    bins=bins,
    labels=labels
)

limit_default = data_clean.groupby("limit_segment")["default payment next month"].mean().reset_index()
print(limit_default)
```

```python
sns.barplot(data=limit_default,
            x="limit_segment",
            y="default payment next month", color="brown")

plt.title("Default Rate by Credit Limit Segment")
plt.xlabel("Credit Limit Segment")
plt.ylabel("Default Rate")
plt.grid(axis="y")
plt.show()
print("_________________________________________________________________________________________________________________________")
```

```python
print("Default Rate vs Debt Ratio")

#tercer análisis, vamos a hallar que tan endeudado está el cliente respecto a su límite de crédito.
#Primero imprimimos medidores estadísticos de la variable debt_ratio
data_clean["DEBT_RATIO"].describe()
```

```python
#Creamos los segmentos 
bins = [0,0.2,0.5,0.8,5]
labels = ["Low Utilization",
          "Medium Utilization",
          "High Utilization",
          "Extreme Utilization"]

data_clean["DEBT_SEGMENT"] = pd.cut(
    data_clean["DEBT_RATIO"],
    bins=bins,
    labels=labels
)
```

```python
debt_default = data_clean.groupby("DEBT_SEGMENT")["default payment next month"].mean()
debt_default = debt_default.reset_index()
debt_default
```

```python
sns.barplot(data=debt_default,
            x="DEBT_SEGMENT",
            y="default payment next month", color="brown")

plt.title("Default Rate by Credit Utilization")
plt.xlabel("Credit Utilization Segment")
plt.ylabel("Default Rate")
plt.xticks(rotation=20)
plt.grid(axis="y")
plt.show()

segment = debt_default.loc[debt_default["default payment next month"].idxmax(),"DEBT_SEGMENT"]
max_value = debt_default["default payment next month"].max()
print(str(segment) + " is the segment with highest default with: " + str(max_value))
print("_____________________________________________________________________________________________________________________________________")
```

```python
print("Payment Behavior vs Default")
#para el cuarto análisis
```

```python
labels = [
    "Very Low Payment",
    "Low Payment",
    "Medium Payment",
    "High Payment"
]

data_clean["PAYMENT_SEGMENT"] = pd.cut(
    data_clean["%_RATIO_PAYT"],
    bins=bins,
    labels=labels
)
```

```python
payment_default = data_clean.groupby("PAYMENT_SEGMENT")["default payment next month"].mean().reset_index()
payment_default
```

```python
plt.figure()
plt.bar(payment_default["PAYMENT_SEGMENT"], payment_default["default payment next month"], color="brown"
)
plt.title("Default Rate by Payment Behavior")
plt.xlabel("Payment Segment")
plt.ylabel("Default Rate")
plt.xticks(rotation=30)
plt.grid(axis="y")
plt.show()

segment_2 = payment_default.loc[payment_default["default payment next month"].idxmax(),"PAYMENT_SEGMENT"]
max_value_2 = payment_default["default payment next month"].max()
print(str(segment_2) + " is the segment with highest default with: " + str(max_value_2))
print("_____________________________________________________________________________________________________________________________________")
```

```python
#Análisis utilizando las edades
bins = [20,30,40,50,60,80]

labels = [
    "20-30",
    "30-40",
    "40-50",
    "50-60",
    "60+"
]
data_clean["age_group"] = pd.cut(
    data_clean["AGE"],
    bins=bins,
    labels=labels
)
```

```python
age_default = data_clean.groupby("age_group")["default payment next month"].mean().reset_index()
age_default
```

```python
sns.barplot(
    data=age_default,
    x="age_group",
    y="default payment next month",
color="brown")
plt.title("Default Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Default Rate")
plt.grid(axis="y")
plt.show()

segment_3 = age_default.loc[age_default["default payment next month"].idxmax(),"age_group"]
max_value_3 = age_default["default payment next month"].max()
print(str(segment_3) + " is the segment with highest default with: " + str(max_value_3))
print("_____________________________________________________________________________________________________________________________________")
```

```python
print("Relation with credit limit, total debt and defaulted")
#gráfico de dispersiónd
no_default = data_clean[data_clean["default payment next month"] == 0]
default = data_clean[data_clean["default payment next month"] == 1]

plt.scatter(
    no_default["LIMIT_BAL"],
    no_default["TOTAL_BILL"],
    label="No Default",
    alpha=0.6,
    s=15
)
plt.scatter(
    default["LIMIT_BAL"],
    default["TOTAL_BILL"],
    label="Default",
    alpha=0.6,
    s=15
)
plt.title("Customer Credit Usage vs Credit Limit")
plt.xlabel("Credit Limit")
plt.ylabel("Total Credit Bill")

plt.legend()
plt.show()
```
